```python
"""
(9)What are the methods for right-justifying, left-justifying, and centering a string?

"""
```

```python
#left adjust
txt = "Ineuron"

x = txt.ljust(30)

print(x)
```

```python
#right adjust
txt = "Ineuron"

x = txt.rjust(30)

print(x)
```

```python
#center 
txt = "Ineuron"

x = txt.center(30)

print(x)
```
